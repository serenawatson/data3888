```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
#import data
keep_countries = ['ARG', 'CHL', 'ECU', 'FRA', 'DEU', 'ITA', 'PER', 'KOR', 'UKR', 'AUS', 'CHN', 'IND', 'NZL', 'NLD', 'CHE', 'ZAF']
pd.reset_option('display')
owid = pd.read_csv("owid-covid-data.csv")
policy = pd.read_csv("covid-vaccination-policy.csv")
owid['population_vaccinated'] = owid['people_vaccinated'] / owid['population'] *100
owid['population_vaccinated'] = owid['population_vaccinated'].fillna(0)
owid_february = owid[owid['date'] == '2021-05-01']
owid_february = owid_february[owid_february['iso_code'].isin(keep_countries)]
owid_february[['iso_code', 'date', 'population_vaccinated']]
```

```python
policy.rename(columns = {'Code':'iso_code'}, inplace = True)
policy_february = policy[policy['Day'] == '2021-05-01']
policy_february = policy_february[policy_february['iso_code'].isin(keep_countries)]
policy_february
```

```python
policy_owid_feb = pd.merge(policy_february, owid_february, on='iso_code')
policy_owid_feb = policy_owid_feb[['iso_code', 'vaccination_policy', 'population_vaccinated']]
policy_owid_feb
```

```python

fig = px.bar(policy_owid_feb, x="iso_code", y="population_vaccinated", color="vaccination_policy", title="May 2021 Policy Vs Vaccinated")
fig.show()
```
